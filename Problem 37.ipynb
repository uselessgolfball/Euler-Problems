

The number 3797 has an interesting property. Being prime itself, it is possible to continuously remove digits from left to right, and remain prime at each stage: 3797, 797, 97, and 7. Similarly we can work from right to left: 3797, 379, 37, and 3.

Find the sum of the only eleven primes that are both truncatable from left to right and right to left.

NOTE: 2, 3, 5, and 7 are not considered to be truncatable primes.


```python
def isPrime(n):
    if n == 1: #1 not prime
        return False
    elif n < 4: #2,3 prime
        return True
    elif n%2 == 0: #evens not prime
        return False
    elif n<9: #odds below 9 are prime
        return True
    elif n%3 == 0: #removes divisible by 3
        return False
    else:
        r = round(n**0.5)
        f = 5
        while f <= r:
            if n%f == 0:
                return False
            if n%(f+2) == 0:
                return False
            f += 6
        return True

primes = []
for i in range(1, 1000000):
    if isPrime(i):
        primes.append(i)
```

```python
primes_1 = []

#remove any even numbers
for num in primes:
    check = 0
    index = 0
    for c in map(int,str(num)):
        if index == 0 and c!=2 and c%2==0:
            check += 1
        elif index > 0 and c%2==0:
            check += 1
        index += 1
    if check == 0:
        primes_1.append(num)
        
print(len(primes),len(primes_1))
```

```python
primes_2 = []

#remove any primes that begin or end with 1
for num in primes_1:
    check = 0
    index = 0
    for c in map(int, str(num)):
        if index == 0 and c == 1:
            check += 1
        elif index == len(str(num))-1 and c == 1:
            check += 1
        index += 1
    if check == 0:
        primes_2.append(num)
        
print(len(primes_1),len(primes_2))
```

```python
print(primes_2)
```

```python
def truncate(N):
    output = []
    for i in range(len(N)-1):
        output.append(N[i+1:])
        output.append(N[:i+1])
    return output

def truncations(n):
    truncations = []
    digits = []
    #pulls digits for set creation
    for c in str(n):
        digits.append(c)
    #generates list
    midput = truncate(digits)
    #returns them to number format
    for item in midput:
        s = ""
        for c in item:
            s += c
        truncations.append(int(s))
    return truncations
    
```

```python
truncations(314)
```

```python
primes_3 = []

for num in primes_2:
    check = 0
    options = truncations(num)
    for i in options:
        if i not in primes:
            check += 1
    if check == 0:
        primes_3.append(num)

print(len(primes_3))
```
