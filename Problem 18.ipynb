```python
def branchCompare(mid, left, right):
    if left >= right:
        return mid + left
    if right > left:
        return mid + right
    
def levelCompare(level_n1, level_n2):
    level_output = []
    n2_count = 0
    for num in level_n1:
        num_out = branchCompare(num, level_n2[n2_count], level_n2[n2_count + 1])
        level_output.append(num_out)
        n2_count += 1
    return level_output
```

```python
def treeCompare(tree):
    level = []
    while len(tree) > 1:
        level = levelCompare(tree[len(tree)-2], tree[len(tree)-1])
        tree.pop()
        tree.pop()
        tree.append(level)
    return tree[0][0]
```
