Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:
Triangle 	  	Tn=n(n+1)/2 	  	1, 3, 6, 10, 15, ...
Pentagonal 	  	Pn=n(3n−1)/2 	  	1, 5, 12, 22, 35, ...
Hexagonal 	  	Hn=n(2n−1) 	  	1, 6, 15, 28, 45, ...

It can be verified that T285 = P165 = H143 = 40755.

Find the next triangle number that is also pentagonal and hexagonal.

```python
# check if it is triangular
def checkTri(n):
    m = quadratic(0.5,0.5,-n)
    for q in m:
        #checks the results of the quadratic
        if q == int(q) and q > 0:
            #if it is a positive integer, then good
            return True
    return False

# check if it is pentagonal
def checkPen(n):
    m = quadratic(1.5,-0.5,-n)
    for q in m:
        if q == int(q) and q > 0:
            return True
    return False

# check if it is hexagonal
def checkHex(n):
    m = quadratic(2,-1,-n)
    for q in m:
        if q == int(q) and q > 0:
            return True
    return False

# solves quadratic
def quadratic(a, b, c):
    x1 = (-b + (b**2-4*a*c)**0.5)/(2*a)
    x2 = (-b - (b**2-4*a*c)**0.5)/(2*a)
    return([x1, x2])
```

```python
n = 286
Hex = n*(2*n-1)

while checkPen(Hex) == False:
    n += 1
    Hex = n*(2*n-1)

print(Hex, n)
```
